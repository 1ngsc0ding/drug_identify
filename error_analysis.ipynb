```python
# ทดสอบเปลี่ยน class ในไฟล์ labels ไล่ทีละไฟล์
ls_txt = []
for a in tqdm(ls_labels):
    path = os.path.join(labels_source,a).replace('\\','/')
    with open(path) as txt:
        txt = txt.read().splitlines()
        # with open('/content/drive/MyDrive/datasets/drug_c3pi/4train/pseudo_lables/'+a, 'w') as dst_file:
        #     for line in txt:
        #         cl, x, y, w, h = [i for i in line.split(' ')]
        #         dst_file.writelines(' '.join(['abc', x, y, w, h])+'\n')
        for line in txt:
            cl, x, y, w, h = [i for i in line.split(' ')]
            ls_txt.append(' '.join([a, cl, x, y, w, h]))


ls_txt
```
